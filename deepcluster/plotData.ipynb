```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('training_metrics_2.csv')
df.head()
```

```python
plt.figure(figsize=(8, 5))
plt.plot(df['Epoch'], df['Loss'], linestyle='-')
plt.xlabel('epochs')
plt.ylabel('Loss over Epochs')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
plt.plot(df['Epoch'], df['NMI'], linestyle='-')
plt.xlabel('epochs')
plt.ylabel('NMI over Epochs')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
plt.plot(df['Epoch'], df['Accuracy'], linestyle='-')
plt.xlabel('epochs')
plt.ylabel('Acc over Epochs')
plt.show()
```
